# Trying to read in the .xml file from the FastA output

```python
import xmltodict
```

```python
#loading the file
with open("BLOSUM50_50.xml","r") as read_file:
    content = read_file.read()
    run1 = xmltodict.parse(content)

print(run1)
```

### Results

```python
print(run1["EBIApplicationResult"]["SequenceSimilaritySearchResult"].keys())
print()
print(run1["EBIApplicationResult"]["SequenceSimilaritySearchResult"]["hits"].keys())
```

```python
# @total is the number of results
print(run1["EBIApplicationResult"]["SequenceSimilaritySearchResult"]["hits"]["@total"])
```

```python
#hit is a list with the results
print(run1["EBIApplicationResult"]["SequenceSimilaritySearchResult"]["hits"]["hit"])
run1_results = run1["EBIApplicationResult"]["SequenceSimilaritySearchResult"]["hits"]["hit"]
```

```python
print(len(run1_results))
```

```python
print(run1_results[0])
```

```python
print(run1_results[0].keys())
```

```python
# everything starting with an @ describes the match sequence
# No of the result, which database it came frome (here SwissProt), 
# the sequence name, the sequence ID in the database, the length of the sequence and a description
print(run1_results[0]['@number'])
print(run1_results[0]['@database'])
print(run1_results[0]['@id'])
print(run1_results[0]['@ac'])
print(run1_results[0]['@length'])
print(run1_results[0]['@description'])
```

```python
# everything starting with an @ describes the match seqeunce
for i in range(len(run1_results)):
    print(run1_results[i]['@number'], run1_results[i]['@database'], run1_results[i]['@id'], run1_results[i]['@ac'], run1_results[i]['@length'], run1_results[i]['@description'])
```

```python
#and finally the alignment or alignments
print(run1_results[0]['alignments'])
print()

print(run1_results[0]['alignments'].keys())
```

```python
# total is again the total number of alignments
# they all have just 1 alignment
numb = ""
for i in range(len(run1_results)):
    numb = numb+run1_results[i]['alignments']["@total"]+" - "
print(numb)
```

```python
print(run1_results[0]['alignments']["alignment"].keys())
```

```python
# @number is the number of the alignment (always1
print(run1_results[0]['alignments']["alignment"]["@number"])
print()
# score is which score, well since its the same as the next I guess it's the optimal alignment score
print(run1_results[0]['alignments']["alignment"]["score"])

# smithWatermanScore is the optimal alignment score
print(run1_results[0]['alignments']["alignment"]["smithWatermanScore"])

# initn is also the optimal alignment score???
print(run1_results[0]['alignments']["alignment"]["initn"])

# init1 is also the optimal alignment score???
print(run1_results[0]['alignments']["alignment"]["init1"])

# opt is also the optimal alignment score???
print(run1_results[0]['alignments']["alignment"]["opt"])
print()
# zScore is the zScore
print(run1_results[0]['alignments']["alignment"]["zScore"])
print()
# bits is the NCBI BLAST+ score, I think sequence similarity?
print(run1_results[0]['alignments']["alignment"]["bits"])
print()

```

```python
# expectation is the E value
print(run1_results[0]['alignments']["alignment"]["expectation"])
print()
# identity is percent identical matches
print(run1_results[0]['alignments']["alignment"]["identity"])
print()
# positives is percent positive matches
print(run1_results[0]['alignments']["alignment"]["positives"])
print()
# overlap is length of the overlap in the alignment
print(run1_results[0]['alignments']["alignment"]["overlap"])
print()
# strand probably applies to nucleotides only
print(run1_results[0]['alignments']["alignment"]["strand"])
print()
# frame???
print(run1_results[0]['alignments']["alignment"]["frame"])
print()
# querySeq is the query sequence (my input)
print(run1_results[0]['alignments']["alignment"]["querySeq"])
print()
# pattern is the alignment pattern
print(run1_results[0]['alignments']["alignment"]["pattern"])
print()
# matchSeq is the target sequence (from the database)
print(run1_results[0]['alignments']["alignment"]["matchSeq"])
print()
```
