# Module 3.3 Exercise 3 - Analysis

```python
import xmltodict
import pandas as pd
```

```python
#loading the file
with open("OIP98664_default.xml","r") as read_file:
    content = read_file.read()
    OIdflt = xmltodict.parse(content)

print(OIdflt)
```

```python
print(OIdflt.keys())
print()
print(OIdflt["BlastOutput"].keys())
```

```python
# the first keys just contain strings with some information on the run and rthe query sequence
print(OIdflt["BlastOutput"]['BlastOutput_program'])
print(OIdflt["BlastOutput"]['BlastOutput_version'])
print(OIdflt["BlastOutput"]['BlastOutput_reference'])
print(OIdflt["BlastOutput"]['BlastOutput_db'])
print(OIdflt["BlastOutput"]['BlastOutput_query-ID'])
print(OIdflt["BlastOutput"]['BlastOutput_query-def'])
print(OIdflt["BlastOutput"]['BlastOutput_query-len'])
print()
# this one has the scoring parameters as well as the expected threshold but not the word size?
print(OIdflt["BlastOutput"]['BlastOutput_param'])
```

```python
# an now it gets interesting:
#print(OIdflt["BlastOutput"]['BlastOutput_iterations'])
print()
print(OIdflt["BlastOutput"]['BlastOutput_iterations'].keys())
print()
print(OIdflt["BlastOutput"]['BlastOutput_iterations']['Iteration'].keys())
```

```python
# Since you can run BLAST several time with some settings I guess this is the iteration of the run
# I only have one interation, so it turns directly into a dictionary. 
# With several iterations I would probably get a list of the iterations

# the first few are again some info in the query (nice that it's here as well) and the iteration number
print(OIdflt["BlastOutput"]['BlastOutput_iterations']['Iteration']['Iteration_iter-num'])
print(OIdflt["BlastOutput"]['BlastOutput_iterations']['Iteration']['Iteration_query-ID'])
print(OIdflt["BlastOutput"]['BlastOutput_iterations']['Iteration']['Iteration_query-def'])
print(OIdflt["BlastOutput"]['BlastOutput_iterations']['Iteration']['Iteration_query-len'])
#the last is a small dictionary about statistics (although I'm not sure what they are or mean)
print(OIdflt["BlastOutput"]['BlastOutput_iterations']['Iteration']['Iteration_stat'])

#the second from last is the acttual hits (see below)
```

```python
# Iteration_hits contains a single key dictionary "Hit" with a list of dictionaries containing the hits
print(OIdflt["BlastOutput"]['BlastOutput_iterations']['Iteration']['Iteration_hits'].keys())
print()
# the length of that list is the number of hits:
print(len(OIdflt["BlastOutput"]['BlastOutput_iterations']['Iteration']['Iteration_hits']["Hit"]))
```

```python
# the information of a single hit can be accessed within the list. 
# the information on the hits are stored in a dictionary
print(OIdflt["BlastOutput"]['BlastOutput_iterations']['Iteration']['Iteration_hits']["Hit"][0].keys())
```

```python
# the first few are again just some info
# the number of the hit, the ID in the respective database, the definition, the accession number and the length of the sequence
print(OIdflt["BlastOutput"]['BlastOutput_iterations']['Iteration']['Iteration_hits']["Hit"][0]['Hit_num'])
print(OIdflt["BlastOutput"]['BlastOutput_iterations']['Iteration']['Iteration_hits']["Hit"][0]['Hit_id'])
print(OIdflt["BlastOutput"]['BlastOutput_iterations']['Iteration']['Iteration_hits']["Hit"][0]['Hit_def'])
print(OIdflt["BlastOutput"]['BlastOutput_iterations']['Iteration']['Iteration_hits']["Hit"][0]['Hit_accession'])
print(OIdflt["BlastOutput"]['BlastOutput_iterations']['Iteration']['Iteration_hits']["Hit"][0]['Hit_len'])

```

```python
# then comes another dictionary with a one-key dictionary with the actual results
print(OIdflt["BlastOutput"]['BlastOutput_iterations']['Iteration']['Iteration_hits']["Hit"][0]['Hit_hsps']['Hsp'].keys())
```

```python
# some numbers and values:
# the nubmer of the Hsp, the max score (bit score), the alignment score, the evalue
# the position of the first and last letters in the alignment of the query sequence and the hit sequence and their frames
# the number of identical letters,  no of positive matches, number of gaps and complete alignment length

print(OIdflt["BlastOutput"]['BlastOutput_iterations']['Iteration']['Iteration_hits']["Hit"][0]['Hit_hsps']['Hsp']['Hsp_num'])
print(OIdflt["BlastOutput"]['BlastOutput_iterations']['Iteration']['Iteration_hits']["Hit"][0]['Hit_hsps']['Hsp']['Hsp_bit-score'])
print(OIdflt["BlastOutput"]['BlastOutput_iterations']['Iteration']['Iteration_hits']["Hit"][0]['Hit_hsps']['Hsp']['Hsp_score'])
print(OIdflt["BlastOutput"]['BlastOutput_iterations']['Iteration']['Iteration_hits']["Hit"][0]['Hit_hsps']['Hsp']['Hsp_evalue'])
print()
print(OIdflt["BlastOutput"]['BlastOutput_iterations']['Iteration']['Iteration_hits']["Hit"][0]['Hit_hsps']['Hsp']['Hsp_query-from'])
print(OIdflt["BlastOutput"]['BlastOutput_iterations']['Iteration']['Iteration_hits']["Hit"][0]['Hit_hsps']['Hsp']['Hsp_query-to'])
print(OIdflt["BlastOutput"]['BlastOutput_iterations']['Iteration']['Iteration_hits']["Hit"][0]['Hit_hsps']['Hsp']['Hsp_hit-from'])
print(OIdflt["BlastOutput"]['BlastOutput_iterations']['Iteration']['Iteration_hits']["Hit"][0]['Hit_hsps']['Hsp']['Hsp_hit-to'])
print(OIdflt["BlastOutput"]['BlastOutput_iterations']['Iteration']['Iteration_hits']["Hit"][0]['Hit_hsps']['Hsp']['Hsp_query-frame'])
print(OIdflt["BlastOutput"]['BlastOutput_iterations']['Iteration']['Iteration_hits']["Hit"][0]['Hit_hsps']['Hsp']['Hsp_hit-frame'])
print()
print(OIdflt["BlastOutput"]['BlastOutput_iterations']['Iteration']['Iteration_hits']["Hit"][0]['Hit_hsps']['Hsp']['Hsp_identity'])
print(OIdflt["BlastOutput"]['BlastOutput_iterations']['Iteration']['Iteration_hits']["Hit"][0]['Hit_hsps']['Hsp']['Hsp_positive'])
print(OIdflt["BlastOutput"]['BlastOutput_iterations']['Iteration']['Iteration_hits']["Hit"][0]['Hit_hsps']['Hsp']['Hsp_gaps'])
print(OIdflt["BlastOutput"]['BlastOutput_iterations']['Iteration']['Iteration_hits']["Hit"][0]['Hit_hsps']['Hsp']['Hsp_align-len'])
```

```python
# and finally the sequences: query, hit and the midline (i.e. the identical and positive matches)
# !! Only the sequence of the aligned part is saved!!!
print(OIdflt["BlastOutput"]['BlastOutput_iterations']['Iteration']['Iteration_hits']["Hit"][0]['Hit_hsps']['Hsp']['Hsp_qseq'])
print()
print(OIdflt["BlastOutput"]['BlastOutput_iterations']['Iteration']['Iteration_hits']["Hit"][0]['Hit_hsps']['Hsp']['Hsp_hseq'])
print()
print(OIdflt["BlastOutput"]['BlastOutput_iterations']['Iteration']['Iteration_hits']["Hit"][0]['Hit_hsps']['Hsp']['Hsp_midline'])
```
