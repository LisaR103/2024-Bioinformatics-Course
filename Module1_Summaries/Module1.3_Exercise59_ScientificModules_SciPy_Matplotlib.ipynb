## SciPy

Scipy is a colletion of mathematical algorithms built on NumPy

The subpackages need to be imported separately, e.g.

    from scipy import linalg, optimize

functions are then used with `.`notation

    linalg.inv(matrix)

A list of subpackages is in the powerpoint, SciPy has its own website: https://scipy.org 

## Matplotlib

A comprehensive library for creating static, animated, and interactive visualizations in Python.    
To use in Python the submodule pyplot is needed:

    import matplotlib.pyplot as plt

some plottypes, all are called with plt.name():
- plot (scatter or line plots)
- bar (vertical bars)
- errorbar (scatter or line plots with errorbars)
- barh (horizontal bars)
- hist (histograms)
- contour (contour plots?
- imshow (heatmap)
- pie
- plot_surface
- bar3d

Modifications similar to ggplot in having them added underneath the call of the plot
- tilte
- xlabel
- ylabel
- xlim
- ylim
- xticks
- yticks
- grid
- legend

```python
import matplotlib.pyplot as plt
```

```python
#some data
xvalues = [1,2,3,4]
yvalues = [10,12,9,11]

# first create the canvas (at least in Jupyter I don't need this)
#plt.figure()
# draw a diagram
plt.plot(xvalues,yvalues)

## modify (this has to be in the same box in Jupyter Notebook!)

#add a title a bit further up than default
plt.title("Lineplot", y = 1.02)
#add Axis Labels
plt.xlabel("Xvalues")
plt.ylabel("Yvalues")
#specifiy the limits of the axes
plt.xlim([0,5])
plt.ylim([5,15])

#etc. etc.
#for legends etc. see powerpoint

# show the plot. I need it in Jupyter too if sth comes after it
plt.show()  

#OR (you cannot do both in Thonny. In Jupyter I can save it an it is shown anyway)

#save the plot in workind directory
#plt.savefig("Image.png")

# close the canvas with (again, not sure I need this in Jupyter. Yes it is if you want todrwa another plot)
plt.close()
```

```python
import random
```

```python
##a Lineplot with at least 10 datapoints
#some data
xvalues1 = list(range(1,11))
yvalues1 = [1,4,9,16,25,36,48,64,81,100]

plt.plot(xvalues1,yvalues1)
#add title
plt.title("A lineplot of exponential values", y = 1.02)
#add Axis Labels
plt.xlabel("Xvalues")
plt.ylabel("Yvalues")
#specifiy the limits of the axes
plt.xlim([1,10])
plt.ylim([1,100])

plt.xticks (xvalues1,["A","B","C","D","E","F","G","H","I","J"])

plt.yticks (yvalues1)
plt.show()
#plt.savefig("Image1.png")
plt.close()
```

```python
##a Scatterplot with at least 500 datapoints
#some data
xvalues2 = list(range(1,1501,3))
yvalues2 = random.sample(list(range(2000)), 500)

plt.plot(xvalues2,yvalues2,'o')
#add title
plt.title("A scatterplot of random values", y = 1.02)
#add Axis Labels
plt.xlabel("Xvalues")
plt.ylabel("Yvalues")
#specifiy the limits of the axes
plt.xlim([1,1500])
plt.ylim([1,2000])

#plt.xticks (xvalues,["A","B","C","D","E","F","G","H","I","J"])

#plt.yticks (yvalues)
plt.show()
#plt.savefig("Image2a.png")
plt.close()
```
